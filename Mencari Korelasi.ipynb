```python
dataset = pd.read_csv('abalone.csv')

print(dataset.head())

print(dataset.info())

print(dataset.describe())
```

Menghapus kolom yang tidak diperlukan.

Kita hanya menggunakan kolom 'Height’, ’ShuckedWeight’, ’WholeWeight’

```python
# data.drop(['Type', 'LongestShell', 'Diameter', 'VisceraWeight',  'ShellWeight',  'Rings'], axis=1, inplace=True)

# print(data.head())
```

```python
dataset.drop(['LongestShell', 'Diameter', 'VisceraWeight',  'ShellWeight',  'Rings'], axis=1, inplace=True)

print(dataset.head())
```

```python
# # Menghitung korelasi Pearson antar kolom
# correlation_matrix_pearson = data.corr(method='pearson')

# correlation_matrix_kendall = data.corr(method='kendall')

# correlation_matrix_spearman = data.corr(method='spearman')
```

```python
class_mapping = {'F': 0, 'I': 1, 'M': 2}
dataset['Type_numeric'] = dataset['Type'].map(class_mapping)

correlation_matrix_pearson = dataset[['Type_numeric', 'Height', 'WholeWeight', 'ShuckedWeight']].corr(method='pearson')
print(correlation_matrix_pearson)

correlation_matrix_kendall = dataset[['Type_numeric', 'Height', 'WholeWeight', 'ShuckedWeight']].corr(method='kendall')
print(correlation_matrix_kendall)

correlation_matrix_spearman = dataset[['Type_numeric', 'Height', 'WholeWeight', 'ShuckedWeight']].corr(method='spearman')
print(correlation_matrix_spearman)
```

```python
# # Menampilkan matriks korelasi dari semua method
# print(correlation_matrix_pearson)

# print(correlation_matrix_kendall)

# print(correlation_matrix_spearman)
```

```python
# Membuat heatmap dari matriks korelasi pearson
plt.figure(figsize=(5, 4))
sns.heatmap(correlation_matrix_pearson, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Heatmap Korelasi Pearson')
plt.show()
```

```python

# Membuat heatmap dari matriks korelasi kendall
plt.figure(figsize=(5, 4))
sns.heatmap(correlation_matrix_kendall, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Heatmap Korelasi Kendall')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Membuat heatmap dari matriks korelasi spearman
plt.figure(figsize=(5, 4))
sns.heatmap(correlation_matrix_spearman, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Heatmap Korelasi Spearman')
plt.show()
```

Dari heatmap diatas bisa dilihat bahwa yang memiliki korelasi kuat (mendekati 1) ialah pasangan data Wholeweight dengan ShuckedWeight
